```python
#INSERTION SORT
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_name = '/home/pc-cs/Desktop/yashasvi1315/DAA/sorting/insertionsortcomparisons.csv'  
df = pd.read_csv(file_name)

# Extract the 'avgcomparisons' and 'nlogn' columns
x = df['input']  # x-axis
avg_comparisons = df['avgcomparisons']  # y-axis
n_log_n = df['nlogn']  # y-axis 

# Plot the data
plt.figure(figsize=(10, 6))

# Plot average comparisons
plt.plot(x, avg_comparisons, label="Average Comparisons", color='blue')

# Plot n log n curve
plt.plot(x, n_log_n, label="n log n", color='red', linestyle='--')

# Add labels, title, and legend
plt.xlabel('Input Size')
plt.ylabel('Value')
plt.title('Comparison between Average Comparisons and n log n')
plt.legend()

plt.grid(True)
plt.show()
```

```python
import csv

# Define your header
header = ['input', 'avgcomparisons', 'nlogn']  # Example header

# Open the existing CSV file
with open('/home/pc-cs/Desktop/yashasvi1315/DAA/sorting/quicksortcomparisons.csv', "r", newline='') as infile:
    # Read the data from the original CSV file
    reader = csv.reader(infile)
    rows = list(reader)

# Open a new CSV file (or the same one) to write the header and data
with open('/home/pc-cs/Desktop/yashasvi1315/DAA/sorting/quicksortcomparisons.csv', "w", newline='') as outfile:
    writer = csv.writer(outfile)
    
    # Write the header first
    writer.writerow(header)
    
    # Write the existing rows from the CSV file
    writer.writerows(rows)

print("Header added successfully!")
```
